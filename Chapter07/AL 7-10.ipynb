### 2) 알고리즘 설명

7-10 : Floyd 알고리즘은 그래프 이론에서 사용되는 알고리즘 중 하나로, 모든 정점 쌍 간의 최단 경로를 찾는 데 사용됩니다. 이 알고리즘은 동적 계획법을 기반으로 하며, 그래프의 모든 정점을 순회하면서 각 정점을 거쳐가는 경로를 비교하여 최단 경로를 업데이트합니다. 이 과정을 반복하여 모든 정점 쌍 간의 최단 경로를 찾습니다.

### 5) 알고리즘 코드

```python
import copy         #리스트의 깊은 복사를 위해 사용
def shortest_path_floyd(vertex, W):     #Floyd의 최단경로 탐색 함수
    vsize = len(vertex)                 #정점의 개수
    D = copy.deepcopy(W)                #깊은 복사
    
    for k in range(vsize) :             #정점 k를 추가할 때
        for i in range(vsize) :         
            for j in range(vsize):      #모든 D[i][j] 갱신
                if (D[i][k] + D[k][j] < D[i][j]):
                    D[i][j] = D[i][k] + D[k][j]
        printD(D)                       #현재 D행렬 출력
```
