# 5주차 실습문제 예제 3.2

### 1) 문제 정의
리스트에 n개의 항목이 들어있다. 이 리스트에서 "탐색키"를 가진 항목을 찾아라. 만약 찾는 항목이 있으면 그 항목의 인덱스를 반환하고, 없으면 -1을 반환하라. 단, 리스트의 항목들은 정렬되어 있지 않다.

### 2) 알고리즘 설명
리스트/ 탐색키를 입력받고 탐색에 성공하면 인덱스를 반환하는 식을 0부터 리스트길이-1까지 반복한다. 탐색에 실패한다면 -1을 반환한다.

### 3) 손으로 푼 예제



### 4) 알고리즘 개요

1) 리스트 'A'와, 탐색키를 입력받는다
2) I가 0부터 len(A)-1까지 반복하는 for문 작성
3) A[i]==key를 활용해 탐색
4) 탐색에 성공하면 인덱스 반환
5) 탐색에 실패하면 -1 반환

### 5) 알고리즘 코드

```python
def selection_sort(A,key):
    for i in range(len(A)):
        if A[i] == key:
            return i
    return -1    
```

### 7) 테스트 코드
