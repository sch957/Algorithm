### 1) 문제 정의
2차원 평면상에 n개의 점이 있다. 가장 인접한 쌍의 거리를 구하라.

### 2) 알고리즘 설명
파이썬의 math모듈의 거듭제곱 연산자를 활용해 각각의 모든 점들의 거리를 비교해서 가장 짧은 점들을 찾는다.

### 4) 알고리즘 개요

1) 2개씩 쌍을 이룬 점들을 입력받는다.
2) 점들의 전체 개수를 n에 저장
3) 이중 for문을 사용해 모든 점들을 비교한다.
4) 유클리드 거리계산 함수를 만들어 계산한다.
5) 기존 거리보다 짧게 나왔다면 최소값변수에 계속 넣어주면서 반복 계산한다.
6) 반복이 끝나면 최소값을 반환한다.

### 5) 알고리즘 코드

```python
def closest_pair(p):
    n = len(p)
    mindist = float("inf")
    for i in range(n-1):
        for j in range(i+1, n):
            dist = distance(p[i], p[j])
            if dist < mindist:
                mindist = dist
    return mindist
```
