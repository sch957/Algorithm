# 5주차 실습문제 예제 3.3

### 1) 문제 정의
길이가 n인 입력 문자열 T와 길이가 m인 패턴 문자열 P가 있다. T에서 가장 먼저 나타나는 P의 위치를 찾아라. 패턴이 없으면 -1을 반환하라.

### 2) 알고리즘 설명
패턴 문자열의 첫 문자와 같은 문자가 있는 인덱스를 찾아 거기서부터 패턴 문자열의 길이만큼 또 비교를 하는 식으로 계속 반복한다.

### 4) 알고리즘 개요

1) 입력 리스트 'T'와, 패턴 리스트 'T'를 입력받는다
2) n과 m에 두 리스트의 길이 저장
3) 0부터 n-m+1까지 반복하는 for문 작성
4) 패턴 문자열을 모두 비교하는 식을 작성
5) 모든 문자가 일치한다면 매칭 성공이니 현재 인덱스 반환
6) 매칭에 실패했다면 -1 반환

### 5) 알고리즘 코드

```python
def string_matching(T,P):
    n = len(T)
    m = len(P)
    for i in range(n-m+1):
        j = 0
        while j<m and P[j] == T[i+j]:
            j = j+1
        if j == m:
            return i
    return -1
```
