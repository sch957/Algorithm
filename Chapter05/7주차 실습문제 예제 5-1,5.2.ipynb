### 1) 문제 정의

병합정렬 알고리즘을 활용하여 두 개의 리스트를 병합정렬하라

### 2) 알고리즘 설명
1.merge_sort(A, left, right)

먼저 리스트의 크기가 1보다 큰지 확인하고, 그렇다면 리스트를 반으로 나누어 merge_sort를 재귀적으로 호출하여 각 부분을 정렬합니다. 그리고 마지막으로 나뉜 부분을 합치는 merge 함수를 호출합니다.

2.merge(A, left, mid, right)
left부터 mid까지의 부분리스트와 mid+1부터 right까지의 부분리스트를 합치는데, 합친 결과를 sorted 배열에 임시로 저장한 후 다시 원래 배열 A에 복사합니다.

### 4) 코드 개요

5.1 
A[left,right]를 오름차순으로 정렬, 항목이 2개 이상인 경우 리스트를 균등 분할 > 부분 리스트 정렬 > 병합

5.2
k = 병합을 위한 임시 리스트의 인덱스, i = 왼쪽 리스트의 인덱스, j = 오른쪽 리스트의 인덱스 
리스트의 슬라이싱을 이용해 임시 리스트로 복사한 후 병합결과를 다시 원래의 리스트 A에 복사하여 

### 5) 알고리즘 코드




```python
def merge_sort(A, left, right) :
    if left < right :
        mid = (left + right) // 2
        merge_sort(A,left,mid)
        merge_sort(A,mid+1,right)
        merge(A,left,mid,right)

def merge(A,left,mid,right):
    k = left
    i = left
    j = mid + 1
    while i <= mid and j <= right :
        if A[i] <= A[j]:
            sorted[k] = A[i]
            i,k = i+1, k+1
        else:
            sorted[k] = A[j]
            j,k = j+1, k+1

    if i > mid :
        sorted[k:k+right-j+1] = A[j:right+1] 
    else :
        sorted[k:k+mid-i+1] = A[i:mid+1]              

    A[left:right+1] = sorted[left:right+1]
```
