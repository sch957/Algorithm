### 2) 알고리즘 설명

알고리즘의 입력은 노드 root로 어떤 트리의 루트를 나타낸다. 만약 루트가 None이면 공백 트리를 의미하므로 높이 0을 바로 반환한다.

### 3) 손으로 푼 예제



### 4) 코드 개요

1행: 루트노드 root  
2행: 공백 트리이면 -> 0을 반환  
3행: 높이가 0 _> 0을 반환  
4행: 왼쪽트리의 높이 -> hLeft  
5행: 오른쪽 트리의 높이 -> hRight  
6행: 더 높은 높이에 1을 더해 반환

### 5) 알고리즘 코드 

```python
def calc_height(root) :
    if root is None :
        return 0
    hLeft = calc_height(root.left)
    hRight = calc_height(root.right)
    return max(hLeft,hRight) + 1
```

### 7) 테스트 코드

```python
class TNode:
    def __init__(self, data, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right

def calc_height(root) :
    if root is None :
        return 0
    hLeft = calc_height(root.left)
    hRight = calc_height(root.right)
    return max(hLeft,hRight) + 1
        
d = TNode('D', None, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('f', None, None)
c = TNode('C', f, None)
root = TNode('A', b, c)
print("트리의 높이=", calc_height(root))
```
