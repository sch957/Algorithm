### 1) 문제 정의

각각 무게가 w이고 가치가 v인 n개의 물건들이 있고, 이것을 배낭에 넣으려고 한다.
이때, 배낭에 넣을 수 있는 용량은 w를 초과하지 않아야 하고, 물건들의 가치의 합이 최대가 되도록 배낭을 채우고, 이때 배낭의 최대가치를 구하라.
단, 물건들은 분할해서 일부분만을 넣을 수도 있다.

### 2) 알고리즘 설명

0-1 배낭 채우기의 알고리즘은 분할해서 채우진 않았지만 이 알고리즘 같은 경우 물건들의 무게를 최대치로 채우고 나서 남은 배낭의 무게를 보고 두번째 가치가 높은 물건을 다 채우는 것이 아닌 분할해서 채우는 알고리즘이다

### 4) 코드 개요

2행 : 내림차순 정렬  
4행 : 전체 배낭의 가치  
6행 : 용량이 다 찬 경우  
7행 : 물건 전체가 들어갈 수 있는 경우  
10행 : 물건의 일부만 넣을 수 있는 경우

### 5) 알고리즘 코드

```python
def knapSack_fractional_greedy(obj, W):
    obj.sort(key = lambda o: o[2]/o[1], reverse=True)
    
    totalValue = 0
    for o in obj :
        if W <= 0 : break
        if W - o[1] >= 0:
            W -= o[1]
            totalValue += o[2]
        else:
            fraction = W / o[1]
            totalValue += o[2] * fraction
            W = int(W - (o[1] * fraction))
            
    return totalValue
```
