### 2) 알고리즘 설명

거스름돈 최소화 알고리즘은 가장 큰 단위의 동전부터 차례로 선택해 남은 금액을 줄여나가 최적의 해를 찾는 방법인 탐욕적 기법을 사용했습니다.

### 3) 손으로 푼 예제



### 4) 코드 개요

2행 : 각 동전의 개수 저장  
4행 : 현재 액면가의 최대 사용 가능 개수  
6행 : 남은 거스름돈 계산

### 5) 알고리즘 코드

```python
def min_coins_greedy(coins,V):
    count = []                      # 각 동전의 개수 저장
    for i in range(len(coins)):
        cnt = V // coins[i]         # 현재 액면가의 최대 사용 가능 개수
        count.append(cnt)
        V -= cnt * coins[i]         # 남은 거스름돈 계산
    return count    
```

### 7) 테스트 코드

```python
# 테스트 코드
coins = [500, 100, 50, 10, 5, 1]
V = 580
print("잔돈 액수 =", V)
print("동전 종류 =", coins)
print("동전 개수 =", min_coins_greedy(coins,V))
```
