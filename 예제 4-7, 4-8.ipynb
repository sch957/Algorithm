# 6주차 실습문제 예제 4-7, 4-8

### 1) 문제 정의

x의 n거듭제곱인 x^n을 계산하라

### 2) 알고리즘 설명

4-7은 반복구조를 이용해 반복문을 이용해 거듭제곱을 구하고, 4-8은 순환구조를 이용하여 함수를 계속 호출해 거듭제곱을 구한다.

### 3) 손으로 푼 예제

### 4) 알고리즘 개요

4-7 
: 반복문을 사용하여 주어진 수 x를 n번 곱하여 x의 n 제곱을 계산합니다.
  각 반복에서는 상수 시간이 소요되므로, n번의 반복이 필요합니다.
  단순하고 직관적인 방법으로 구현되어 있지만, 입력 크기가 커질수록 비효율적입니다.
  
4-8
: 재귀적으로 문제를 절반씩 분할하며, 지수 n을 절반으로 나누어 재귀적으로 호출합니다.ㅋ
  홀수인 경우와 짝수인 경우를 나누어 처리하여 재귀 호출을 최적화합니다.
  입력 크기가 커져도 효율적으로 동작하며, 특히 지수 n이 큰 경우에 더욱 효과적입니다.

### 5) 알고리즘 코드

```python
def slow_power(x,n):            # 예제 4-7
    result = 1.0
    for i in range(n):          # 루프 n번 반복
        result = result * x
    return result     
```

```python
def power(x,n):                 # 예제 4-8
    if n == 0:                        # 종료 조건
        return 1                     # 모든 수의 0승은 1
    elif (n%2)==0:                   # n이 짝수
        return power(x*x,n//2)       # 정수의 나눗셈
    else :                           # n이 홀수
        return x*power(x*x,(n-1)//2)  
```
