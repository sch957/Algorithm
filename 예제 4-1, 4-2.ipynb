### 1) 문제 정의

하향식 축소 정복 기법을 이용한 팩토리얼 알고리즘

### 2) 알고리즘 설명

4-1 
: 함수가 호출되면, 먼저 n이 1인지 확인합니다. 만약 n이 1이라면, 팩토리얼 값이 1이므로 1을 반환합니다.
  그렇지 않은 경우, n과 factorial_recur(n-1)의 곱을 반환합니다. 이 부분이 재귀 호출이 일어나는 부분입니다. 즉, n의 팩토리얼을 계산하기     위해 n과 n-1의 팩토리얼 값을 곱하고, n-1의 팩토리얼 값을 구하기 위해 다시 factorial_recur 함수를 호출합니다.

4-2
: 1부터 n까지의 모든 정수를 곱하여 결과를 반환한다.
  반복문을 통해 상향식 접근 방법으로 구현되어 있다.

### 3) 손으로 푼 예제

### 4) 알고리즘 개요

4-1
: 재귀적으로 팩토리얼을 계산합니다.
  n이 1이 되면 재귀 호출을 멈추고 1을 반환합니다.
  그렇지 않은 경우, 현재 n과 n-1의 팩토리얼을 곱하여 반환합니다.
  재귀 호출을 통해 n부터 1까지의 정수를 모두 곱하여 팩토리얼을 계산합니다.

4-2
: 반복문을 사용하여 팩토리얼을 계산합니다.
  1부터 n까지의 정수를 반복하여 곱하여 결과를 반환합니다.
  초기값으로 1을 설정하고, 각 반복에서는 현재까지의 곱에 다음 정수를 곱하여 결과를 갱신합니다.  

### 5) 알고리즘 코드

```python
# 예제 4-1
def factorial_recur(n):                     # 축소 정복 기법 : 하향식 접근
    if n == 1 : 
        return 1
    else :
        return (n*factorial_recur(n-1))    # 순환구조 이용     
```

```python
# 예제 4-2
def factorial_iter(n):                     # 축소 정복 기법 : 상향식 접근
    result = 1
    for k in range(1,n+1):                # 반복구조 이용  
        result = result * k
    return result   
```

### 7) 테스트 코드

```python
print(factorial_recur(5)) 
print(factorial_recur(3)) 
print(factorial_recur(10)) 

print(factorial_iter(5)) 
print(factorial_iter(3)) 
print(factorial_iter(10)) 
```
